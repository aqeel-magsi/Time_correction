```python
#this code will helps us to plot the time difference between predicted and picked arrivals time. input files are from
#bash script with name "get picks for each station.sh"
import matplotlib.dates as dates
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats
import matplotlib.pyplot as pls 
```

```python
#read the local data files
H33=pd.read_csv("local_shz_H33.dat",sep=' ', names=["day","date","dist","td"])
H34=pd.read_csv("local_shz_H34.dat",sep=' ', names=["day","date","dist","td"])
H36=pd.read_csv("local_shz_H36.dat",sep=' ', names=["day","date","dist","td"])
K06=pd.read_csv("local_shz_K06.dat",sep=' ', names=["day","date","dist","td"])
K20=pd.read_csv("local_shz_K20.dat",sep=' ', names=["day","date","dist","td"])
```

```python
#read the tele data files
tH33=pd.read_csv("tele_shz_H33.dat",sep=' ', names=["day","date","dist","td"])
tH34=pd.read_csv("tele_shz_H34.dat",sep=' ', names=["day","date","dist","td"])
tH36=pd.read_csv("tele_shz_H36.dat",sep=' ', names=["day","date","dist","td"])
tK06=pd.read_csv("tele_shz_K06.dat",sep=' ', names=["day","date","dist","td"])
tK20=pd.read_csv("tele_shz_K20.dat",sep=' ', names=["day","date","dist","td"])
```

```python
import datetime
t1_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in t1]
t2_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in t2]
t3_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in t3]
t4_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in t4]
t5_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in t5]
#print(x_values)
```

```python
#for tele files
import datetime
tt1_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in tt1]
tt2_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in tt2]
tt3_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in tt3]
tt4_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in tt4]
tt5_values = [datetime.datetime.strptime(d,"%Y/%m/%d").date() for d in tt5]
#print(x_values)
len(tt1_values)
len(tt2_values)
len(tt3_values)
```

```python
import pandas as pd
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import matplotlib.pylab as plt
from matplotlib.dates import MonthLocator, DateFormatter
from matplotlib.dates import DateFormatter, WeekdayLocator, MonthLocator, DayLocator, MONDAY
import matplotlib.dates as mdates
import numpy as np
#plt.subplot_tool()


fig =plt.figure(figsize=(12.5,12.5))
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})

#subplots axis
#ax3=fig.add_subplot(3,1,3)
ax1=fig.add_subplot(2,1,1)
ax1.tick_params(labelbottom=False,which='both')
ax2=fig.add_subplot(2,1,2,sharey=ax1)
#ax2.tick_params(labelbottom=False,which='both')
#ax2.tick_params(labelbottom=False)

#ax3.tick_params(labelbottom=False)
ax1.plot(tt1_values,ta,'o', color='b',label="H33",markersize=8)
ax1.plot(tt2_values,tb,'o', color='g',label="H34",markersize=8)
ax1.plot(tt3_values,tc,'o', color='r',label="H36",markersize=8)
ax1.plot(tt4_values,td,'o', color='c',label="K06",markersize=8)
ax1.plot(tt5_values,te,'o', color='k',label="K20",markersize=8)
ax1.axhline(y=0,color='red',linestyle='-', linewidth = 2.0)
#ax1.legend(loc='lower right',ncol=len(df.columns))
ax1.legend(loc="upper center",bbox_to_anchor=(0.5, 1.1),ncol=5,facecolor="w")
ax1.axhspan(-20,-15, facecolor='0.5', alpha=0.5, lw=0)
#ax2.set_title('(B) Time difference for each Local Events')
#ax2.set_xlabel('Date')
ax1.set_ylabel('$T_{pred}$' '$- T_{obsr}$ ')
ax1.set_ylim([-30, 15])
ax1.set_yticks([-30,-20, -10, 0, 10])

#plot 01

ax2.plot(t1_values,a,'o', color='b',label="H33",markersize=8)
ax2.plot(t2_values,b,'o', color='g',label="H34",markersize=8)
ax2.plot(t3_values,c,'o', color='r',label="H36",markersize=8)
ax2.plot(t4_values,d,'o', color='c',label="K06",markersize=8)
ax2.plot(t5_values,e,'o', color='k',label="K20",markersize=8)
ax2.axhline(y=0,color='red',linestyle='-', linewidth = 2.0)
#ax2.legend(loc='best')
#ax1.set_title('(A) Time difference for each Local Events')
#ax1.set_xlabel('Date')
ax2.axhspan(-24,-11, facecolor='0.5', alpha=0.5, lw=0)

ax2.set_ylabel('$T_{pred}$' '$- T_{obsr}$ ')
ax2.set_ylim([-30, 15])

ax2.xaxis.set_major_locator(plt.MaxNLocator(5))

date_form = DateFormatter("%d/%m/y")
loc = mdates.MonthLocator(interval=2)
ax2.xaxis.set_minor_locator(dates.MonthLocator())
#ax2.xaxis.set_minor_formatter(dates.DateFormatter('%b\n%Y'))
ax2.xaxis.set_major_locator(loc)
fmt = mdates.DateFormatter('%b\n%y')
ax2.xaxis.set_major_formatter(fmt)

ax1.annotate("(a) Teleseismic events", xy=(0.01, 0.05), xycoords="axes fraction",fontsize = 18.0, backgroundcolor='white',alpha=1)
ax2.annotate("(b) Local events", xy=(0.01, 0.05), xycoords="axes fraction",fontsize = 18.0, backgroundcolor='white')
fig.tight_layout(pad=3.0)
#fig.tight_layout()
plt.savefig('Differential_time.jpeg')
plt.savefig('Differential_time.pdf')
```
